```python
import matplotlib.pyplot as plt
from random_walk import RandomWalk

active = True
while active:
    # RandomWalk
    rw = RandomWalk(50000) 
    rw.fill_walk()

    point_numbers = list(range(rw.num_points))
    plt.scatter(rw.x_values, rw.y_values, s=2,c=point_numbers,cmap=plt.cm.Blues)
    # 
    plt.scatter(0,0,s=10,color=(0,1,0))
    plt.scatter(rw.x_values[-1],rw.y_values[-1],s=10,color=(1,0,0))
    plt.show()

    keep_running = input("Make another rank walk?(y/n): ")
    if keep_running == "n":
        active = False
```
